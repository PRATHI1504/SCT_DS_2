```python
# Step 1: Import Required Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Step 2: Load the Dataset
file_path = '/content/2019.csv'
data = pd.read_csv(file_path)
```

```python
# Step 3: Display the First Few Rows
print("Dataset Preview:")
print(data.head())
```

```python
# Step 4: Check for Missing Values
print("\nMissing Values:")
print(data.isnull().sum())
```

```python
# Step 5: Exploratory Data Analysis (EDA)
# Histogram for 'Score'
plt.figure(figsize=(10, 6))
sns.histplot(data['Score'], kde=True, color='blue', bins=15)
plt.title("Distribution of Happiness Scores")
plt.xlabel("Happiness Score")
plt.ylabel("Frequency")
plt.show()
```

```python
# Boxplot for GDP per Capita
plt.figure(figsize=(10, 6))
sns.boxplot(x=data['GDP per capita'])
plt.title("Boxplot of GDP per Capita")
plt.xlabel("GDP per Capita")
plt.show()
```

```python
# Scatterplot between GDP and Happiness Score
plt.figure(figsize=(10, 6))
sns.scatterplot(x=data['GDP per capita'], y=data['Score'], hue=data['Country or region'])
plt.title("GDP per Capita vs Happiness Score")
plt.xlabel("GDP per Capita")
plt.ylabel("Happiness Score")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
# Step 7: Correlation Matrix
# Select only numerical columns
numerical_data = data.select_dtypes(include=['float64', 'int64'])

# Calculate and plot the correlation matrix
plt.figure(figsize=(10, 6))
correlation_matrix = numerical_data.corr()
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Matrix")
plt.show()
```

```python
# Step 8: Insights
print("\nInsights:")
print("1. The histogram shows the distribution of happiness scores across countries.")
print("2. The scatterplot indicates a positive relationship between GDP per capita and happiness score.")
print("3. The correlation matrix provides insights into relationships between numerical variables.")
```
